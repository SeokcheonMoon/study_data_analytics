## pandas.concat()
- 기능 : 같은 컬럼으로 붙임

```python
df_BRAZIL = pd.read_csv("BRAZIL.csv")
df_Canada = pd.read_csv("Canada.csv")
df_FRANCE = pd.read_csv("FRANCE.csv")
df_ITALY = pd.read_csv("ITALY.csv")
df_India = pd.read_csv("India.csv")
df_USA = pd.read_csv("USA.csv")
```

```python
df_BRAZIL.shape, df_Canada.shape, df_FRANCE.shape ,df_ITALY.shape ,df_India.shape ,df_USA.shape
```

```python
df_merged = pd.concat(objs = [df_BRAZIL,df_Canada,df_FRANCE,df_ITALY,df_India,df_USA])
df_merged
```
