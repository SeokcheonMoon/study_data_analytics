```python
import pandas as pd
```

```python
double_list = [
    [1000, '과자', '2019-12-31', '반품'],
    [2000, '음료', '2020-03-02', '정상'],
    [3000, '아이스크림', '2020-02-03', '정상'],
    [1000, '과자', '2019-12-31', '반품'],
    [4000, '초콜릿', '2020-03-15', '정상'],
    [5000, '사탕', '2020-04-01', '정상'],
    [6000, '빵', '2020-05-20', '반품'],
    [7000, '우유', '2020-06-30', '정상'],
    [8000, '치즈', '2020-07-07', '반품']
]
double_columns = ['가격','종류','판매일자','반품여부']
df_sales = pd.DataFrame(data=double_list, columns=double_columns)
df_sales.info()
```

### pd.pivot_table()
- data : DataFrame
- index : group by (unique())
- values : aggregation의 대상 column 이름
- aggfunc : 집계(aggregation)

```python
df_sales
```

```python
# 각 그룹별 row 갯수
pd.pivot_table(data = df_sales[["가격", "종류", "반품여부"]], index= "반품여부", values = "가격", aggfunc = "count")
```

```python
pd.pivot_table(data = df_sales[["가격", "종류", "반품여부"]], index= "반품여부", values = "가격", aggfunc = "sum") # sum 말고 mean이나 median도 가능함.
```

### pivot_table with numpy

```python
import numpy as np
```

```python
df_sales_pivots = pd.pivot_table(data = df_sales[["가격", "종류", "반품여부"]], index= "반품여부", values = "가격", aggfunc = np.sum)
df_sales_pivots
```

```python
# df_sales_pivots = df_sales_pivots.reset_index()
df_sales_pivots.reset_index(inplace=True)       # 위 아래 둘다 같은 의미.
```
