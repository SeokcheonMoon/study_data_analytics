## 3. 'Embarked' 열에서 결측치가 있는 행을 찾고, 가장 많이 등장하는 항구로 채우기.

```python
df_titan.describe(include=object)         # 가장많은 빈도수의 항구 찾기
```

```python
df_titan_embarked = df_titan["Embarked"].fillna("S")          # 가장많은 빈도수의 항구로 채우기
```

```python
df_titan_embarked.isnull().sum()       #Embarked 값이 다 채워짐
```

## 4. 'Fare' 열에서 평균보다 3 표준편차(mean + 3 * std) 이상 높은 이상치를 가진 행들을 찾기.



```python
df_titan.describe()
```

```python
condition_mean = df_titan['Fare'].mean()
condition_mean
```

```python
condition_3std = df_titan['Fare'].std() * 3
condition_3std
```

```python
condition_meanstd = f"Fare >= {condition_mean}+{condition_3std}"
df_titan.query(condition_meanstd)
```

## 5. 'SibSp' (형제 자매/배우자 수)가 5 이상인 이상치를 가진 행들을 찾기.

```python
condition_sibsp = "SibSp >= 5"
df_titan.query(condition_sibsp)
```

## 6. 'Parch' (부모/자녀 수) 열에서 0이 아닌 값을 가진 행들의 평균 'Age' 계산하기.


```python
condition_parch = "Parch != 0"
df_titan_condition_parch = df_titan.query(condition_parch)
df_titan_condition_parch
```

```python
df_titan_condition_parch["Age"].mean()
```
