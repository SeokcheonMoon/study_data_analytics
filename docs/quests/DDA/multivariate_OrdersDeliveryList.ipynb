```python
# 3. 2019년 1월 7일 이전에 배송이 시작된 주문들을 찾으세요.
```

```python
def convert_time(x):                                            # x -> "2019-01-11 오후 05:32" -> 2019-01-11 17:32 로 전환되는 과정의 function
    date, ampm, time = x.split()
    if ampm == "오후" and time != "12:00":
        hour, minute = map(int, time.split(":"))
        time = f"{hour+12 if hour != 12 else hour}:{minute}"
    return pd.to_datetime(f"{date} {time}", format="%Y-%m-%d %H:%M")

df_shopmall["배송시작일_datetime"]=df_shopmall["배송시작일"].apply(convert_time)
df_shopmall["배송시작일_datetime"]
```

```python
df_shopmall.query("배송시작일_datetime<'2019-01-07 00:00:00'")
```

```python
# 4. 배송 시작일과 배송 완료일이 3일 이상 차이 나는 주문들을 찾으세요.
```

```python
def convert_time(x):                                            # x -> "2019-01-11 오후 05:32" -> 2019-01-11 17:32 로 전환되는 과정의 function
    date, ampm, time = x.split()
    if ampm == "오후" and time != "12:00":
        hour, minute = map(int, time.split(":"))
        time = f"{hour+12 if hour != 12 else hour}:{minute}"
    return pd.to_datetime(f"{date} {time}", format="%Y-%m-%d %H:%M")

df_shopmall["배송시작일_datetime"]=df_shopmall["배송시작일"].apply(convert_time)
df_shopmall["배송완료일_datetime"]=df_shopmall["배송완료일"].apply(convert_time)

df_shopmall["배송소요일"] = df_shopmall["배송완료일_datetime"]-df_shopmall["배송시작일_datetime"]
df_shopmall["배송소요일"]
```

```python
condition_morethan_3day = df_shopmall["배송소요일"]
condition_morethan_3day
```

```python
df_shopmall.query(condition_morethan_3day >= "3 days")
```
